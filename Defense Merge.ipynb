```python
teams, def_rnk, def_rtg, season = [], [], [], []
for filename in os.listdir('./defense_scrape/csv'):
    with open(f'./defense_scrape/csv/{filename}', 'r') as f:
        lines = f.read().split('\n')
        n = len(lines[1:])
        def_rnk+=[line.split(',')[0] for line in lines[1:] if len(line) > 0]
        teams+=[line.split(',')[1] for line in lines[1:] if len(line) > 0]
        def_rtg+=[line.split(',')[6] for line in lines[1:] if len(line) > 0]
        season+=[filename.split('.')[0] for i in range(n)]
defense = pd.DataFrame({'Team': teams, 'Season': season, 'Def_Rnk': def_rnk, 'Def_Rtg': def_rtg})
defense['Team_Abb'] = defense.Team.map(teams_map)
defense.to_pickle('./defense_scrape/defense.pkl')
```
